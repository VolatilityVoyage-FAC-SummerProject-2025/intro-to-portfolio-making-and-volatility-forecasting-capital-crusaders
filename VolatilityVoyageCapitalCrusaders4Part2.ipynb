```python
returns = data['Close'].pct_change()
rolling_vol = returns.rolling(window=20).std()
```

Three Forecast Methods

```python
def rolling_mean(rollig_vol,window=5): #method 1
    return rollig_vol.rolling(window).mean().dropna()
def ewma(returns:pd.Series,l=0.94): #method 2
    return (((returns**2).ewm(alpha=1-l).mean())**0.5).dropna()
def ar_1(rolling_vol:pd.Series): #method 3
    model = AutoReg(rolling_vol.dropna(), lags=1).fit() 
    return model.predict(start=len(data), end=len(data)) 

def normalise(arr:pd.Series):
    return (arr-arr.min())/(arr.max()-arr.min())
```

```python
def compute_rmse(true, predicted):
    return np.sqrt(np.mean((true - predicted) ** 2))

def find_best_forecast(rolling_vol, returns):
    forecasts = {
        'Rolling Mean': rolling_mean(rolling_vol),
        'EWMA': ewma(returns),
        'AR(1)': ar_1(rolling_vol)
    }

    # Align forecast length to actual rolling vol for fair RMSE comparison
    aligned_vol = rolling_vol.dropna().iloc[-len(forecasts['Rolling Mean']):]
    rmses = {
        method: compute_rmse(aligned_vol, forecast.iloc[-len(aligned_vol):])
        for method, forecast in forecasts.items()
    }

    best_method = min(rmses, key=rmses.get)
    print(f"Best method: {best_method} with RMSE: {rmses[best_method]:.6f}")
    return forecasts[best_method]
```

```python
def strat(df: pd.DataFrame):
    df = df.copy()
    df['ema14'] = df['Close'].ewm(span=14).mean()
    df['atr'] = (df['High'] - df['Low']).rolling(window=14).mean()

    df['signal'] = np.where((df['Close'] > df['ema14']) & (df['Close'].shift(1) <= df['ema14'].shift(1)), 1, 0)
    df['signal'] = np.where((df['Close'] < df['ema14']) & (df['Close'].shift(1) >= df['ema14'].shift(1)), -1, df['signal'])
    return df

# --- Backtesting for Multiple Assets ---
def backtest_multiple(data_dict, weights, signal_col='signal', initial_capital=1_000_000):
    capital_per_stock = {ticker: initial_capital * weight for ticker, weight in weights.items()}
    all_trades = {}
    all_daily = {}

    for ticker, df in data_dict.items():
        capital = capital_per_stock[ticker]
        df = df.copy()
        df['Signal'] = df[signal_col]
        position = 0
        cash = capital
        trades = []
        shares_held = 0
        entry_price = None
        atr = None
        stop_loss = None
        entry_date = None
        portfolio_value = []

        for date, row in df.iterrows():
            price = row['Close']
            signal = row['Signal']
            atr = row['atr']

            if position == 1 and price < stop_loss:
                signal = -1  # Trigger stop-loss exit

            if signal == 1 and position == 0:
                shares_to_buy = int(cash / price)
                invested = shares_to_buy * price
                cash -= invested
                shares_held = shares_to_buy
                entry_price = price
                entry_date = date
                stop_loss = price - 1.5 * atr
                position = 1

            elif signal == -1 and position == 1:
                pnl = (price - entry_price) * shares_held
                cash += shares_held * price
                trades.append({
                    'Entry Date': entry_date,
                    'Exit Date': date,
                    'Entry Price': entry_price,
                    'Exit Price': price,
                    'Shares': shares_held,
                    'PnL': pnl
                })
                shares_held = 0
                entry_price = None
                stop_loss = None
                position = 0
                entry_date = None

            current_value = cash + (shares_held * price if position == 1 else 0)
            portfolio_value.append({
                'Date': date,
                'Portfolio Value': current_value
            })

        all_trades[ticker] = pd.DataFrame(trades)
        all_daily[ticker] = pd.DataFrame(portfolio_value).set_index('Date')

    return all_trades, all_daily

def compute_metrics_all(trade_dict, daily_dict):
    all_metrics = {}

    for ticker in trade_dict:
        df = trade_dict[ticker]
        daily_df = daily_dict[ticker]
        if df.empty or daily_df.empty:
            continue

        r = daily_df['Portfolio Value'].iloc[-1] - daily_df['Portfolio Value'].iloc[0]
        rp = (r * 100) / daily_df['Portfolio Value'].iloc[0]

        sharpe = np.nan
        if 'Portfolio Value' in daily_df:
            returns = daily_df['Portfolio Value'].pct_change().dropna()
            if returns.std() != 0:
                sharpe = (returns.mean() / returns.std()) * np.sqrt(252)

        all_metrics[ticker] = {
            'Return': r,
            'Return %': rp,
            'Sharpe Ratio': sharpe,
            'Total Trades': len(df),
            'Winning Trades': (df['PnL'] > 0).sum(),
            'Losing Trades': (df['PnL'] <= 0).sum()
        }

    return pd.DataFrame(all_metrics).T
```

```python
best_forecast_func = find_best_forecast(rolling_vol.mean(axis=1), returns.mean(axis=1))
risk_score = float(input("Enter risk averseness (0 = high risk appetite, 1 = risk averse): "))
normalized_vol = normalise(best_forecast_func)
latest_vol = rolling_vol.iloc[-1]  # latest vol per asset
distance = abs(normalise(latest_vol) - risk_score)
top3_tickers = distance.nsmallest(3).index.tolist()
print("Top 3 stocks matching risk profile:", top3_tickers)
```

```python
std_devs = data['Close'][top3_tickers].pct_change().std()
weights = 1 / std_devs
weights /= weights.sum()
weights_dict = weights.to_dict()
weights_dict
```

```python
data_dict = dict([(x,strat(data.xs(x,axis=1, level=1))) for x in top3_tickers])
trades,daily=backtest_multiple(data_dict,weights_dict)
metrics = compute_metrics_all(trades,daily)
display(metrics)
```
